## Borra algunas filas

```python
df.drop(list(range(5,35)), axis = 'index')
```

##Â Borra algunas columnas

```python
df.drop(['location','date', 'perc_sequences'], axis = 'columns')
```

## Escribe una funciÃ³n y aplÃ­cala a alguna de las columnas

```python
def Check_omicron(x):
    return x == 'Omicron'
df["omicron_check"] = df.variant.apply(Check_omicron)
df.sample(4)
```

##Â Muestra los valores Ãºnicos de algunas columnas

```python
df.num_sequences.unique()
```

##Â Filtra los datos por algunos de los valores que hay dentro de alguna columna

```python
df2 = df[df["location"] == "Canada"]
df2.sample(3) # We take in acount the cases in Canada. 
```

##Â Repite el paso anterior con condicionales

```python
df3 = df[(df.location == "Canada") & (df.omicron_check == True)] # only the omicron cases in canada
df3.info() # only 44 registered
```

## Elige una columna y ponla como index

```python
df4 = df.set_index("date")
df4.sample(3)
```

##Â Resetea el index

```python
df5 = df.reset_index()
df5.sample(3)
```
